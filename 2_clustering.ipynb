```python
cluster_analysis = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Cluster']].groupby('Cluster').mean()
cluster_analysis
```

# Reporte de Recomendaciones
### Estrategia Digital y Visibilidad
La presencia en línea debe trascender la simple oferta académica para posicionar a HJybridge como una institución moderna y flexible. La clave está en capitalizar las tendencias algorítmicas. El modelo de búsqueda actual dicta que la universidad no solo debe aparecer en búsquedas directas de su rubro central **(tecnología y educación)**, sino también incursionar en otros géneros y audiencias.

Es fundamental realizar un análisis profundo del algoritmo de las principales plataformas **(Google, redes sociales)** para identificar palabras clave de cola larga que atraigan a públicos no directamente interesados en tecnología a priori, pero que sí valoran la flexibilidad de horarios y la oportunidad de finalizar una carrera universitaria de forma más expedita. La estrategia de contenido debe incluir publicaciones y anuncios dirigidos a estos intereses transversales. Por ejemplo, en lugar de solo promocionar **"Ingeniería de Software"**, se debe promocionar la **"Gestión de tu tiempo y tu futuro profesional"**, usando la tecnología y la flexibilidad como los principales atractivos. La inversión en SEO **(Search Engine Optimization) y SEM (Search Engine Marketing)** debe recalcar estos beneficios diferenciadores para garantizar la máxima visibilidad.
### Marketing Experiencial y Práctico
La organización de **Competencias o Hackatones presenciales** es una recomendación estratégica de alto valor para el marketing experiencial y la generación de leads cualificados. Estos eventos sirven como vitrina de la infraestructura de la universidad y como un filtro natural para atraer a personas realmente comprometidas con la tecnología.

Se recomienda establecer un calendario anual de al menos dos (2) Hackatones. La promoción debe ser intensa y dirigida a estudiantes de bachillerato y universitarios de otras instituciones. Ofrecer premios significativos y la validación de la experiencia mediante certificados o, incluso, becas parciales, motivará la asistencia. Este evento, además de ser una herramienta de captación, funcionará como generador de contenido de alta calidad (videos, fotos y testimonios) para todas las campañas digitales futuras, reforzando la imagen de HJybridge como un hub de innovación
### Diferienciación Cultural y Prestigio Académico
La universidad debe formalizar una **nueva cultura interna y académica** que se centre en la exigencia y la simulación del entorno laboral de alta presión. Es decir, dejar de ser simplemente una universidad y transformarse en un **Centro de Desarrollo de Talento de Alto Rendimiento.**

Esto se logra mediante la implementación de **modelos y estructuras de aprendizaje agresivas**. Esto incluye la organización constante de competencias internas, proyectos de desarrollo de software con plazos estrictos y el manejo de proyectos complejos que fuercen a los estudiantes a mejorar su resiliencia y adaptabilidad. Esta filosofía no es un obstáculo, sino un argumento de venta. El mensaje de marketing debe ser claro: HJybridge prepara a sus egresados para **liderar y manejar la presión** en cualquier empresa tecnológica, lo cual justifica la inversión educativa y diferencia a sus egresados en el mercado laboral.
### Fomento de la tutoría y el Aprendizaje Continuo
Finalmente, se recomienda instituir un programa de Cursos Extra curriculares avalados por la universidad, que sean impartidos por los propios alumnos más avanzados o destacados.
La promoción de estos cursos debe ir de la mano con la oferta principal, atrayendo tanto a alumnos internos como a profesionales externos interesados en upskilling o especialización puntual.
